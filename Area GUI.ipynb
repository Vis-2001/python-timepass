```python
from tkinter import*
import math
root=Tk(className="Different Shapes!")
root.configure(bg="black")
root1=Tk(className="Dimensions")
root1.configure(bg="silver")

def square():
    l=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the side of the square")
    Lba.grid(row=1,column=0)
    l.grid(row=1,column=1)
    def abc():
        x=float(l.get())
        area=x**2
        Lb1 = Label(root1, text = "Area of square is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=3,column=0)
        Lb2.grid(row=3,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=2,column=0,columnspan=2)
    root1.mainloop()
    
def rectangle():
    l=Entry(root1,width=10)
    b=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the length of the rectangle") 
    Lbb=Label(root1,text="Enter the breadth of the rectangle")
    Lba.grid(row=1,column=0)
    l.grid(row=1,column=1)
    Lbb.grid(row=2,column=0)
    b.grid(row=2,column=1)
    def abc():
        x=float(l.get())
        y=float(b.get())
        area=x*y
        Lb1 = Label(root1, text = "Area of rectangle is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=4,column=0)
        Lb2.grid(row=4,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=3,column=0,columnspan=2)
    root1.mainloop()

def triangle():
    a=Entry(root1,width=10)
    b=Entry(root1,width=10)
    c=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the side 1 of the triangle") 
    Lbb=Label(root1,text="Enter the side 2 of the triangle") 
    Lbc=Label(root1,text="Enter the side 3 of the triangle") 
    Lba.grid(row=1,column=0)
    Lbb.grid(row=2,column=0)
    Lbc.grid(row=3,column=0)
    a.grid(row=1,column=1)
    b.grid(row=2,column=1)
    c.grid(row=3,column=1)
    def abc():
        x=float(a.get())
        y=float(b.get())
        z=float(c.get())
        s=(x+y+z)/2
        g=s*(s-x)*(s-y)*(s-z)
        area=math.sqrt(g)
        Lb1 = Label(root1, text = "Area of triangle is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=5,column=0)
        Lb2.grid(row=5,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=4,column=0,columnspan=2)
    root1.mainloop()

def parallelogram():
    b=Entry(root1,width=10)
    h=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the base of the parallelogram") 
    Lbb=Label(root1,text="Enter the height of the parallelogram") 
    Lba.grid(row=1,column=0)
    Lbb.grid(row=2,column=0)
    b.grid(row=1,column=1)
    h.grid(row=2,column=1)
    def abc():
        x=float(b.get())
        y=float(h.get())
        area=x*y
        Lb1 = Label(root1, text = "Area of parallelogram is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=4,column=0)
        Lb2.grid(row=4,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=3,column=0,columnspan=2)
    root1.mainloop()

def Trapezoid():
    b1=Entry(root1,width=10)
    b2=Entry(root1,width=10)
    h=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the length of first parallel side")
    Lbb=Label(root1,text="Enter the length of second parallel side")
    Lbc=Label(root1,text="Enter the distance between the parallel sides")
    Lba.grid(row=1,column=0)
    Lbb.grid(row=2,column=0)
    Lbc.grid(row=3,column=0)
    b1.grid(row=1,column=1)
    b2.grid(row=2,column=1)
    h.grid(row=3,column=1)
    def abc():
        x=float(b1.get())
        y=float(b2.get())
        z=float(h.get())
        area=((x+y)/2)*z
        Lb1 = Label(root1, text = "Area of trapezoid is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=5,column=0)
        Lb2.grid(row=5,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=4,column=0,columnspan=2)
    root1.mainloop()

def circle():
    r=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the radius of the circle")
    Lba.grid(row=1,column=0)
    r.grid(row=1,column=1)
    def abc():
        x=float(r.get())
        area=3.14*(x**2)
        Lb1 = Label(root1, text = "Area of circle is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=3,column=0)
        Lb2.grid(row=3,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=2,column=0,columnspan=2)
    root1.mainloop()

def ellipse():
    a=Entry(root1,width=10)
    b=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the length of the major axis")
    Lbb=Label(root1,text="Enter the length of the minor axis")
    Lba.grid(row=1,column=0)
    Lbb.grid(row=2,column=0)
    a.grid(row=1,column=1)
    b.grid(row=2,column=1)
    def abc():
        x=float(a.get())
        y=float(b.get())
        area=3.14*x*y
        Lb1 = Label(root1, text = "Area of ellipse is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=4,column=0)
        Lb2.grid(row=4,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=3,column=0,columnspan=2)
    root1.mainloop()

def cube():
    a=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the side of the cube")
    Lba.grid(row=1,column=0)
    a.grid(row=1,column=1)
    def abc():
        x=float(a.get())
        area=6*(x**2)
        Lb1 = Label(root1, text = "Area of cube is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=3,column=0)
        Lb2.grid(row=3,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=2,column=0,columnspan=2)
    root1.mainloop()

def cuboid():
    l=Entry(root1,width=10)
    b=Entry(root1,width=10)
    h=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the length of the cuboid")
    Lbb=Label(root1,text="Enter the breadth of the cuboid")
    Lbc=Label(root1,text="Enter the height of the cuboid")
    Lba.grid(row=1,column=0)
    Lbb.grid(row=2,column=0)
    Lbc.grid(row=3,column=0)
    l.grid(row=1,column=1)
    b.grid(row=2,column=1)
    h.grid(row=3,column=1)
    def abc():
        x=float(l.get())
        y=float(b.get())
        z=float(h.get())
        area=2*((x*y)+(y*z)+(z*x))
        Lb1 = Label(root1, text = "Area of cuboid is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=5,column=0)
        Lb2.grid(row=5,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=4,column=0,columnspan=2)
    root1.mainloop()

def cur_cylinder():
    r=Entry(root1,width=10)
    h=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the radius of the cylinder")
    Lba.grid(row=1,column=0)
    Lbb=Label(root1,text="Enter the height of the cylinder")
    Lbb.grid(row=2,column=0)
    r.grid(row=1,column=1)
    h.grid(row=2,column=1)
    def abc():
        x=float(r.get())
        y=float(h.get())
        area=2*3.14*x*y
        Lb1 = Label(root1, text = "Curved surface area of cylinder is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=4,column=0)
        Lb2.grid(row=4,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=3,column=0,columnspan=2)
    root1.mainloop()

def tot_cylinder():
    r=Entry(root1,width=10)
    h=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the radius of the cylinder")
    Lba.grid(row=1,column=0)
    Lbb=Label(root1,text="Enter the height of the cylinder")
    Lbb.grid(row=2,column=0)
    r.grid(row=1,column=1)
    h.grid(row=2,column=1)
    def abc():
        x=float(r.get())
        y=float(h.get())
        area=2*3.14*x*(x+y)
        Lb1 = Label(root1, text = "Total surface area of cylinder is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=4,column=0)
        Lb2.grid(row=4,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=3,column=0,columnspan=2)
    root1.mainloop()

def cur_cone():
    r=Entry(root1,width=10)
    l=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the radius of the cone")
    Lba.grid(row=1,column=0)
    Lbb=Label(root1,text="Enter the slant length of the cone")
    Lbb.grid(row=2,column=0)
    r.grid(row=1,column=1)
    l.grid(row=2,column=1)
    def abc():
        x=float(r.get())
        y=float(l.get())
        area=3.14*x*y
        Lb1 = Label(root1, text = "Curved surface area of cone is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=4,column=0)
        Lb2.grid(row=4,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=3,column=0,columnspan=2)
    root1.mainloop()

def tot_cone():
    r=Entry(root1,width=10)
    l=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the radius of the cone")
    Lba.grid(row=1,column=0)
    Lbb=Label(root1,text="Enter the slant length of the cone")
    Lbb.grid(row=2,column=0)
    r.grid(row=1,column=1)
    l.grid(row=2,column=1)
    def abc():
        x=float(r.get())
        y=float(l.get())
        area=3.14*x*(y+x)
        Lb1 = Label(root1, text = "Total surface area of cone is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=4,column=0)
        Lb2.grid(row=4,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=3,column=0,columnspan=2)
    root1.mainloop()

def sphere():
    r=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the radius of the sphere")
    Lba.grid(row=1,column=0)
    r.grid(row=1,column=1)
    def abc():
        x=float(r.get())
        area=4*3.14*(x**2)
        Lb1 = Label(root1, text = "Area of sphere is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=3,column=0)
        Lb2.grid(row=3,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=2,column=0,columnspan=2)
    root1.mainloop()

def hemisphere():
    r=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the radius of the hemisphere")
    Lba.grid(row=1,column=0)
    r.grid(row=1,column=1)
    def abc():
        x=float(r.get())
        area=3*3.14*(x**2)
        Lb1 = Label(root1, text = "Area of hemisphere is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=3,column=0)
        Lb2.grid(row=3,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=2,column=0,columnspan=2)
    root1.mainloop()

def sector():
    r=Entry(root1,width=10)
    a=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the radius of the sector")
    Lba.grid(row=1,column=0)
    Lbb=Label(root1,text="Enter the angle covered by the sector in radians")
    Lbb.grid(row=2,column=0)
    r.grid(row=1,column=1)
    a.grid(row=2,column=1)
    def abc():
        x=float(r.get())
        y=float(a.get())
        area=((x**2)*y)/2
        Lb1 = Label(root1, text = "Area of sector is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=4,column=0)
        Lb2.grid(row=4,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=3,column=0,columnspan=2)
    root1.mainloop()

def pyramid():
    b=Entry(root1,width=10)
    l=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the length of the base of the pyramid")
    Lba.grid(row=1,column=0)
    Lbb=Label(root1,text="Enter the slant length of the pyramid")
    Lbb.grid(row=2,column=0)
    b.grid(row=1,column=1)
    l.grid(row=2,column=1)
    def abc():
        x=float(b.get())
        y=float(l.get())
        area=(2*x*y) + (x**2)
        Lb1 = Label(root1, text = "Area of pyramid is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=4,column=0)
        Lb2.grid(row=4,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=3,column=0,columnspan=2)
    root1.mainloop()

def prism():
    l=Entry(root1,width=10)
    b=Entry(root1,width=10)
    h=Entry(root1,width=10)
    s=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the length of the prism")
    Lba.grid(row=1,column=0)
    Lbb=Label(root1,text="Enter the length of the base of triangle of prism")
    Lbb.grid(row=2,column=0)
    Lbc=Label(root1,text="Enter the height of the triangle of the prism")
    Lbc.grid(row=3,column=0)
    Lbd=Label(root1,text="Enter the length of the side of triangle of prism")
    Lbd.grid(row=4,column=0)
    l.grid(row=1,column=1)
    b.grid(row=2,column=1)
    h.grid(row=3,column=1)
    s.grid(row=4,column=1)
    def abc():
        w=float(l.get())
        x=float(b.get())
        y=float(h.get())
        z=float(s.get())
        area=(x*y) + (2*w*z) +(w*x)
        Lb1 = Label(root1, text = "Area of isosceles prism is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=6,column=0)
        Lb2.grid(row=6,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=5,column=0,columnspan=2)
    root1.mainloop()

def rhombus():
    d1=Entry(root1,width=10)
    d2=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the length of first diagonal")
    Lba.grid(row=1,column=0)
    Lbb=Label(root1,text="Enter the length of second diagonal")
    Lbb.grid(row=2,column=0)
    d1.grid(row=1,column=1)
    d2.grid(row=2,column=1)
    def abc():
        x=float(d1.get())
        y=float(d2.get())
        area=(x*y)/2
        Lb1 = Label(root1, text = "Area of rhombus is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=4,column=0)
        Lb2.grid(row=4,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=3,column=0,columnspan=2)
    root1.mainloop()

def cur_hollow_cylinder():
    R=Entry(root1,width=10)
    h=Entry(root1,width=10)
    r=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the outer radius of the cylinder")
    Lba.grid(row=1,column=0)
    Lbb=Label(root1,text="Enter the height of the cylinder")
    Lbb.grid(row=2,column=0)
    Lbc=Label(root1,text="Enter the inner radius of the cylinder")
    Lbc.grid(row=3,column=0)
    R.grid(row=1,column=1)
    h.grid(row=2,column=1)
    r.grid(row=3,column=1)
    def abc():
        x=float(R.get())
        y=float(h.get())
        z=float(r.get())
        area=(2*3.14*x*y)+(2*3.14*z*y)
        Lb1 = Label(root1, text = "Curved surface area of hollow cylinder is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=5,column=0)
        Lb2.grid(row=5,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=4,column=0,columnspan=2)
    root1.mainloop()

def tot_hollow_cylinder():
    R=Entry(root1,width=10)
    h=Entry(root1,width=10)
    r=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the outer radius of the cylinder")
    Lba.grid(row=1,column=0)
    Lbb=Label(root1,text="Enter the height of the cylinder")
    Lbb.grid(row=2,column=0)
    Lbc=Label(root1,text="Enter the inner radius of the cylinder")
    Lbc.grid(row=3,column=0)
    R.grid(row=1,column=1)
    h.grid(row=2,column=1)
    r.grid(row=3,column=1)
    def abc():
        x=float(R.get())
        y=float(h.get())
        z=float(r.get())
        area=(2*3.14*x*y)+(2*3.14*z*y)+(2*3.14*((x**2)-(z**2)))
        Lb1 = Label(root1, text = "Total surface area of hollow cylinder is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=5,column=0)
        Lb2.grid(row=5,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=4,column=0,columnspan=2)
    root1.mainloop()

def regular_polygon():
    n=Entry(root1,width=10)
    s=Entry(root1,width=10)
    Lba=Label(root1,text="Enter the number of sides of the polygon")
    Lba.grid(row=1,column=0)
    Lbb=Label(root1,text="Enter the side of the polygon")
    Lbb.grid(row=2,column=0)
    n.grid(row=1,column=1)
    s.grid(row=2,column=1)
    def abc():
        x=float(n.get())
        y=float(s.get())
        c=abs(math.tan(3.14/x))
        d=1/(4*c)
        area=((y**2)*x)*d
        Lb1 = Label(root1, text = "Area of regular polygon is:")
        Lb2 = Label(root1, text = area,bg="gold")
        Lb1.grid(row=4,column=0)
        Lb2.grid(row=4,column=1)
    bta=Button(root1,text="Click here to get the answer!",command=abc)
    bta.grid(row=3,column=0,columnspan=2)
    root1.mainloop()

bt1 = Button(root, text = "Square", command = square,bg="yellow",width=40)
bt2 = Button(root, text = "Rectangle", command = rectangle,bg="yellow",width=40)
bt3 = Button(root, text = "Triangle", command = triangle,bg="yellow",width=40)
bt4 = Button(root, text = "Parallelogram", command = parallelogram,bg="yellow",width=40)
bt5 = Button(root, text = "Trapezoid", command = Trapezoid,bg="yellow",width=40)
bt6 = Button(root, text = "Circle", command = circle,bg="yellow",width=40)
bt7 = Button(root, text = "Ellipse", command = ellipse,bg="yellow",width=40)
bt8 = Button(root, text = "Cube", command = cube,bg="yellow",width=40)
bt9 = Button(root, text = "Cuboid", command = cuboid,bg="yellow",width=40)
bt10 = Button(root, text = "Cylinder(Curved Surface Area)", command = cur_cylinder,bg="yellow",width=40)
bt11 = Button(root, text = "Cylinder(Total Surface Area)", command = tot_cylinder,bg="yellow",width=40)
bt12 = Button(root, text = "Cone(Curved Surface Area)", command = cur_cone,bg="yellow",width=40)
bt13 = Button(root, text = "Cone(Total Surface Area)", command = tot_cone,bg="yellow",width=40)
bt14 = Button(root, text = "Sphere", command = sphere,bg="yellow",width=40)
bt15 = Button(root, text = "Hemisphere", command = hemisphere,bg="yellow",width=40)
bt16 = Button(root, text = "Sector", command = sector,bg="yellow",width=40)
bt17 = Button(root, text = "Pyramid", command = pyramid,bg="yellow",width=40)
bt18 = Button(root, text = "Isoceles Prism", command = prism,bg="yellow",width=40)
bt19 = Button(root, text = "Rhombus", command = rhombus,bg="yellow",width=40)
bt20 = Button(root, text = "Hollow Cylinder(Curved Surface Area)", command = cur_hollow_cylinder,bg="yellow",width=40)
bt21 = Button(root, text = "Hollow Cylinder(Total Surface Area)", command = tot_hollow_cylinder,bg="yellow",width=40)
bt22 = Button(root, text = "Regular Polygon", command = regular_polygon,bg="yellow",width=40)

bt1.grid(row=1,column=0,columnspan=2)
bt2.grid(row=2,column=0,columnspan=2)
bt3.grid(row=3,column=0,columnspan=2)
bt4.grid(row=4,column=0,columnspan=2)
bt5.grid(row=5,column=0,columnspan=2)
bt6.grid(row=6,column=0,columnspan=2)
bt7.grid(row=7,column=0,columnspan=2)
bt8.grid(row=8,column=0,columnspan=2)
bt9.grid(row=9,column=0,columnspan=2)
bt10.grid(row=10,column=0,columnspan=2)
bt11.grid(row=11,column=0,columnspan=2)
bt12.grid(row=12,column=0,columnspan=2)
bt13.grid(row=13,column=0,columnspan=2)
bt14.grid(row=14,column=0,columnspan=2)
bt15.grid(row=15,column=0,columnspan=2)
bt16.grid(row=16,column=0,columnspan=2)
bt17.grid(row=17,column=0,columnspan=2)
bt18.grid(row=18,column=0,columnspan=2)
bt19.grid(row=19,column=0,columnspan=2)
bt20.grid(row=20,column=0,columnspan=2)
bt21.grid(row=21,column=0,columnspan=2)
bt22.grid(row=22,column=0,columnspan=2)
root.mainloop()
```
