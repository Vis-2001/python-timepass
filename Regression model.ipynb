```python
df=pd.read_csv("C:\\Users\\unuiw\\OneDrive\\Desktop\\sp.csv")
```

```python
df.head(10)
```

```python
fem=list()
male=list()
for i in df.gender:
    if i=='female':
        fem.append(1)
        male.append(0)
    else:
        fem.append(0)
        male.append(1)
```

```python
df['female']=fem
df['male']=male
```

```python
df=df.drop('gender',axis=1)
```

```python
dum=['race/ethnicity','parental level of education','lunch','test preparation course']
df=pd.get_dummies(df,columns=dum)
```

```python
df.head()
```

```python
y1=df[['math score','reading score','writing score']]
x=df.drop(['math score','reading score','writing score'],axis=1)
```

```python
from sklearn.model_selection import train_test_split as tts 
```

```python
from sklearn.linear_model import LogisticRegression as LR
```

```python
y1.head()
```

```python
y=df.sum(axis=1)
```

```python
y.head()
```

```python
pas=300*0.7
result=list()
for i in y:
    if i>=pas:
        result.append(1)
    else:
        result.append(0)
```

```python
yhat=pd.DataFrame(result)
```

```python
yhat.head()
```

```python
x_train,x_test,y_train,y_test=tts(x,yhat,test_size=0.3)
```

```python
model= LR()
```

```python
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.score(x_test,y_test)
```
