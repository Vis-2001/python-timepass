```python
import math
def square(l):
    area=l**2
    return area
def rectangle(l,b):
    area=l*b
    return area
def triangle(a,b,c):
    s=(a+b+c)/2
    x=s*(s-a)*(s-b)*(s-c)
    area=math.sqrt(x)
    return area
def parallelogram(b,h):
    area=b*h
    return area
def Trapezoid(b1,b2,h):
    area=((b1+b2)/2)*h
    return area
def circle(r):
    area=3.14*(r**2)
    return area
def ellipse(a,b):
    area=3.14*a*b
    return area
def cube(a):
    area=6*(a**2)
    return area
def cuboid(l,b,h):
    area=2*((l*b)*(b*h)*(h*l))
    return area
def cur_cylinder(r,h):
    area=2*3.14*r*h
    return area
def tot_cylinder(r,h):
    area=2*3.14*r*(r+h)
    return area
def cur_cone(r,l):
    area=3.14*r*l
    return area
def tot_cone(r,l):
    area=3.14*r*(r+l)
    return area
def sphere(r):
    area=4*3.14*(r**2)
    return area
def hemisphere(r):
    area=3*3.14*(r**2)
    return area
def sector(r,a):
    area=((r**2)*a)/2
    return area
def pyramid(b,l):
    area=(2*b*l) + (b**2)
    return area
def prism(l,b,h,s):
    area=(b*h) + (2*l*s) +(l*b)
    return area
def rhombus(d1,d2):
    area=(d1*d2)/2
    return area
def cur_hollow_cylinder(R,h,r):
    area=(2*3.14*R*h)+(2*3.14*r*h)
    return area
def tot_hollow_cylinder(R,h,r):
    area=(2*3.14*R*h)+(2*3.14*r*h)+(2*3.14*((R**2)-(r**2)))
    return area
def regular_polygon(n,s):
    c=math.tan(180/n)
    area=n*c*0.5
    return area
while True:
    print("Welcome to Area Calculator")
    print("""
    1.Square
    2.Rectangle
    3.Triangle
    4.Parallelogram
    5.Trapezoid
    6.Circle
    7.Ellipse
    8.Cube
    9.Cuboid
    10.Cylinder(Curved Surface Area)
    11.Cylinder(Total Surface Area)
    12.Cone(Curved Surface Area)
    13.Cone(Total Surface Area)
    14.Sphere
    15.Hemisphere
    16.Sector
    17.Pyramid
    18.Isoceles prism
    19.Rhombus
    20.Hollow Cylinder(Curved Surface Area)
    21.Hollow Cylinder(Total Surface Area)
    22.Regular polygon""")
    ch=int(input("Enter choice:"))
    if ch==1:
        l=float(input("Enter the side of the square:"))
        ar=square(l)
        print("Area of the given square=",ar)
    elif ch==2:
        l=float(input("Enter the length of the rectangle:"))
        b=float(input("Enter the breadth of the rectangle:"))
        ar=rectangle(l,b)
        print("Area of the given rectangle=",ar)
    elif ch==3:
        a=float(input("Enter side 1 of the triangle:"))
        b=float(input("Enter side 2 of the triangle:"))
        c=float(input("Enter side 3 of the triangle:"))
        ar=triangle(a,b,c)
        print("Area of the given triangle=",ar)
    elif ch==4:
        b=float(input("Enter the base of the parallelogram:"))
        h=float(input("Enter the height of the parallelogram:"))
        ar=parallelogram(b,h)
        print("Area of the given parallelogram=",ar)
    elif ch==5:
        b1=float(input("Enter the length of the first parallel side:"))
        b2=float(input("Enter the length of the second parallel side:"))
        h=float(input("Enter the distance between the parallel sides:"))
        ar=Trapezoid(b1,b2,h)
        print("Area of the given trapezium=",ar)
    elif ch==6:
        r=float(input("Enter the radius of the circle:"))
        ar=circle(r)
        print("Area of the given circle=",ar)
    elif ch==7:
        a=float(input("Enter the length of the major axis:"))
        b=float(input("Enter the length of the minor axis:"))
        ar=ellipse(a,b)
        print("Area of the given ellipse=",ar)
    elif ch==8:
        a=float(input("Enter the length of the side of the cube:"))
        ar=cube(a)
        print("Area of the given cube=",ar)
    elif ch==9:
        l=float(input("Enter the length of the cuboid:"))
        b=float(input("Enter the breadth of the cuboid:"))
        h=float(input("Enter the height of the cuboid:"))
        ar=cuboid(l,b,h)
        print("Area of the given cuboid=",ar)
    elif ch==10:
        r=float(input("Enter the radius of the cylinder:"))
        h=float(input("Enter the height of the cylinder:"))
        ar=cur_cylinder(r,h)
        print("Curved surface area of the given cylinder=",ar)
    elif ch==11:
        r=float(input("Enter the radius of the cylinder:"))
        h=float(input("Enter the height of the cylinder:"))
        ar=tot_cylinder(r,h)
        print("Total surface area of the given cylinder=",ar)
    elif ch==12:
        r=float(input("Enter the radius of the cone:"))
        l=float(input("Enter the slant length of the cone:"))
        ar=cur_cone(r,l)
        print("Curved surface area of the given cone=",ar)
    elif ch==13:
        r=float(input("Enter the radius of the cone:"))
        l=float(input("Enter the slant length of the cone:"))
        ar=tot_cone(r,l)
        print("Total surface area of the given cone=",ar)
    elif ch==14:
        r=float(input("Enter the radius of the sphere:"))
        ar=sphere(r)
        print("Area of the given sphere=",ar)
    elif ch==15:
        r=float(input("Enter the radius of the hemishpere:"))
        ar=hemisphere(r)
        print("Area of the given hemisphere=",ar)
    elif ch==16:
        r=float(input("Enter the radius of the sector:"))
        a=float(input("Enter the angle covered by the sector:"))
        ar=sector(r,a)
        print("Area of the given sector:",ar)
    elif ch==17:
        b=float(input("Enter the length of the base of the pyramid:"))
        l=float(input("Enter the slant length of the pyramid:"))
        ar=pyramid(b,l)
        print("Area of the given pyramid:",ar)
    elif ch==18:
        l=float(input("Enter the length of the prism:"))
        b=float(input("Enter the length of base of triangle of prism:"))
        h=float(input("Enter the height of the trangle of prism:"))
        s=float(input("Enter the length of side of triangle of prism:"))
        ar=prism(l,b,h,s)
        print("Area of the given prism=",ar)
    elif ch==19:
        d1=float(input("Enter the length of the first diagnol:"))
        d2=float(input("Enter the length of the second diagnol"))
        ar=rhombus(d1,d2)
        print("Area of the given rhombus=",ar)
    elif ch==20:
        R=float(input("Enter the outer radius of the cylinder:"))
        h=float(input("Enter the height of the cylinder:"))
        r=float(input("Enter the inner radius of the cylinder:"))
        ar=cur_hollow_cylinder(R,h,r)
        print("Curved surface area of the given hollow cylinder=",ar)
    elif ch==21:
        R=float(input("Enter the outer radius of the cylinder:"))
        h=float(input("Enter the height of the cylinder:"))
        r=float(input("Enter the inner radius of the cylinder:"))
        ar=tot_hollow_cylinder(R,h,r)
        print("Total surface area of the given hollow cylinder=",ar)
    elif ch==22:
        n=int(input("Enter the number of sides of the regular polygon:"))
        s=float(input("Enter the length of side of the regular polygon:"))
        ar=regular_polygon(n,s)
        print("Area of the given regular polygon=",ar)
    else:
        print("Wrong input")
    print("Enter 1 to calculate area of more shapes")
    print("Enter any number other than 1 to exit")
    chr=int(input("Enter choice:"))
    if chr==1:
        continue
    else:
        break
```
